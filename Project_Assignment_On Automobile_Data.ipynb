## details of all Toyota cars

```python
toyota_car_detail = data.loc[(data['company']=='toyota'), 
                             ['company', 'body-style', 'wheel-base', 'length', 'engine-type', 
                              'num-of-cylinders', 'horsepower','average-mileage', 'price']]
toyota_car_detail
```

## total cars per company

```python
data['company'].value_counts()
```

## The car of each company which costs highest among other cars, and the average mileage providing car company name

```python
group_car = data.groupby('company')
high_price = group_car['company', 'price'].max()
high_price
```

```python
group_car = data.groupby('company')
average_mileage = group_car['company', 'average-mileage'].mean()
average_mileage
```

## All companies producing the cars with 'num-of-cylinders'is equal to'six

```python
cylinders_six = data.loc[(data['num-of-cylinders'] == 'six'), ['company', 'num-of-cylinders']]
cylinders_six
```

## All comapnies details which are producing the cars with 'average-mileage' greater than or equal to 35 and number of cylinders equal to four.

```python
mark1 = data['average-mileage'] >=35
mark2 = data['num-of-cylinders'] == 'four'

new_data = data.loc[(mark1 & mark2), ['company', 'average-mileage', 'num-of-cylinders']]
new_data
```
